```python
# rename the No-show column as Output_label
df = df.rename(columns = {"No-show": "Output_label"})
```

```python
# convert no and yes as 0 and 1
df.Output_label = df.Output_label.replace({'No': 0, 'Yes': 1}) # or df.replace(('Yes','No'),(1,0))
```

```python
df.Gender = df.Gender.replace({'F':0,'M':1})
```

```python
df = df.rename(columns = {"City ": "City"})
```

```python
df.City = df.City.replace({'No':0, 'Yes':1})
```

```python
# Now calculate the prevalence of the positive class
#Make function
def cal_prevalence(y_actual):
    p = sum(y_actual)/len(y_actual)
    return p
```

```python
# Find the unique values
for c in list(df.columns):
    n = df[c].unique()
    # if number of unique values is less then 30, print the values, else print numbers
    if len(n)<30:
        print(c)
        print(n)
    else :
        print(c + ":", len(n), 'Unique Values')
```

```python
# In our data we have ? / so we want to replace '?' with Nan values
import numpy as np
df = df.replace('?', np.nan)
```

```python
# convert to the date format
df["ScheduledDay"] = pd.to_datetime(df["ScheduledDay"], format = "%Y-%m-%d %H:%M:%S")
df["AppointmentDay"] = pd.to_datetime(df["AppointmentDay"], format = "%Y-%m-%d %H:%M:%S")
df.head()
```

```python
# Now find the difference and create new columns
df["Num_days"] = (df["AppointmentDay"] - df["ScheduledDay"]).dt.days
df.head()
```

```python
# convert - in 0
df.Num_days = np.where(df.Num_days<0, 0, df.Num_days)
df.head()
```

```python
df["day_of_week"] = df["ScheduledDay"].dt.dayofweek
df["month"] = df["ScheduledDay"].dt.month
df["week"] = df["ScheduledDay"].dt.week
```

```python
col_cat = ["Gender", 
           "PRIORITY_TYPE_CODE", 
           "Index of Multiple Deprivation Decile", 
           "Income Decile", 
           "Employment Decile", 
           "Education and Skills Decile", 
           "Health and Disability Decile"]
df_cat = pd.get_dummies(df[col_cat])
```

```python
# Join the Data with 
col_data = df[col_names]
new_df = pd.concat([col_data,df_cat], axis = 1)
```

```python
cols_all_cat = list(df_cat.columns)
print("Total Number of Features:", len(col_names + cols_all_cat))
```

```python
x = new_df
y = df["Output_label"]
```

```python
df.keys()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
# Apply K-Nearest-Neighbours
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=100)
knn.fit(x_train, y_train)
```

```python
# Apply Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state = 0)
lr.fit(x_train, y_train)
```

```python
# Decision Tree classifier
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(random_state = 0)
tree.fit(x_train, y_train)
```

```python
# Apply Random forest classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 0)
rf.fit(x_train, y_train)
```

```python
# Print Accuracy
from sklearn.metrics import accuracy_score, precision_score, recall_score
def print_Report(model_name, y_test, y_pred):
    print("Accuracy of", model_name, accuracy_score(y_test, y_pred))
    print("Recall of", model_name, recall_score(y_test, y_pred))
    print("Precision of", model_name, precision_score(y_test, y_pred))
```

```python
# K-Nearest-Neighbours
y_pred = knn.predict(x_test)
# Print Prediction Acuuracy
print_Report("KNeighborsClassifier", y_test, y_pred)
cm = metrics.confusion_matrix(y_test, y_pred)
cm
```

```python
# Logistic regression
y_pred = lr.predict(x_test)
# Print Prediction Acuuracy
print_Report("Logistic Regression", y_test, y_pred)
cm = metrics.confusion_matrix(y_test, y_pred)
cm
```

```python
# Decision tree
y_pred = tree.predict(x_test)
# Print Prediction Acuuracy
print_Report("Decision Trees", y_test, y_pred)
cm = metrics.confusion_matrix(y_test, y_pred)
cm
```

```python
# Random Forest
y_pred = rf.predict(x_test)
# Print Prediction Acuuracy
print_Report("Random Forest", y_test, y_pred)
cm = metrics.confusion_matrix(y_test, y_pred)
cm
```

```python
df_results = pd.DataFrame({"Classifier": ["K-NN", "LR", "DT", "RD"], 
                           "Accuracy": [accuracy_score(y_test, knn.predict(x_test)),
                                        accuracy_score(y_test, lr.predict(x_test)),
                                        accuracy_score(y_test, tree.predict(x_test)),
                                        accuracy_score(y_test, rf.predict(x_test))],
                          "Recall": [recall_score(y_test, knn.predict(x_test)),
                                     recall_score(y_test, lr.predict(x_test)),
                                     recall_score(y_test, tree.predict(x_test)),
                                     recall_score(y_test, rf.predict(x_test))],
                          "Precision": [precision_score(y_test, knn.predict(x_test)),
                                       precision_score(y_test, lr.predict(x_test)),
                                       precision_score(y_test, tree.predict(x_test)),
                                       precision_score(y_test, rf.predict(x_test))]})
```
