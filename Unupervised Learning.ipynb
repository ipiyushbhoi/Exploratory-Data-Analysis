```python
from sklearn.cluster import KMeans 

#WE dont want private col,so we will drop it

new_df=df.drop("Private",axis=1)
```

```python
#Here , no Train Test Split is used,as this is unsupervised learning
km=KMeans(n_clusters=2) #for dividing the data into two clusters("Comparing with yes/no")
km.fit(new_df)
```

```python
#Factor Importance
```

```python
#Now we make Converter for cluster
```

```python
def Converter(cluster):
    if(cluster=="Yes"):
        return 1
    else:
        return 0
```

```python
#Now, we create new column as 'Cluster'

df["Cluster"]=df["Private"].apply(Converter)   #PRIVATE IS THE ACTUAL COLUMN, & WE ARE CONVERTING IT INTO 1 & 0 FORM
```

```python
from sklearn import metrics                    
cm=metrics.confusion_matrix(df["Cluster"],df["Private"].apply(Converter)) 
cm
```

```python
#cluster to be plotted
```

```python
pred=km.labels_   #(LABELS CONTAIN PREDICTED VALUES)
```

```python
cm=metrics.confusion_matrix(df["Cluster"],pred) 
```

```python
cm
```

```python
sns.heatmap(cm,annot=True)
plt.show()
```
